```python
import matplotlib.patches as patches

img = mpimg.imread('face1.jpg')

fig, ax = plt.subplots(figsize = (10,10)) #테두리 프레임, 진짜사진
ax.imshow(img)

rect_face = patches.Rectangle((x,y),w,h,
                             linewidth = 5,
                             edgecolor = 'r',
                             facecolor = 'none')
ax.add_patch(rect_face)

plt.text(10,400, annotation, wrap=True, fontsize = 17, color ='white')
plt.show()
```

```python
import matplotlib.patches as patches

img = mpimg.imread('face2.jpg')

fig, ax = plt.subplots(figsize = (10,10)) #테두리 프레임, 진짜사진
ax.imshow(img)


for each in parsed['faces']: #인원수대로
    x,y,w,h = each['roi'].values()
    gender,gen_confi = each['gender'].values()
    emotion,emo_confi = each['emotion'].values()
    age,age_confi = each['age'].values()
    
    rect_face = patches.Rectangle((x,y),w,h,
                             linewidth = 5,
                             edgecolor = 'r',
                             facecolor = 'none')
    ax.add_patch(rect_face)
    annotation = gender + " : " + str(round(gen_confi,2)*100)+'%' + '\n' + emotion + " : " + str(round(emo_confi,2)*100) +'%'+ '\n' + age + " : " + str(round(age_confi,2)*100)+'%'  
    plt.text(x,y, annotation, wrap=True, fontsize = 13, color ='blue')

    
plt.show()
```
